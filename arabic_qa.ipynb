```python
import pandas as pd
import numpy as np
```

```python
from transformers import pipeline
```

```python
df = pd.read_csv("/content/arabic_qa_sample.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
```

```python
df = df.dropna().drop_duplicates().reset_index(drop=True)
df
```

```python
qa_pipeline = pipeline("question-answering", model="deepset/xlm-roberta-base-squad2")
```

```python
sample_context = df.loc[0, "context"]
sample_question = df.loc[0, "question"]

result = qa_pipeline(question=sample_question, context=sample_context)
result
```

```python
print("Question:", sample_question)
print("Context:", sample_context)
print("Predicted Answer:", result["answer"])
print("Expected Answer:", df.loc[0, "answer"])
```

```python
predictions = []

for _, row in df.iterrows():
    result = qa_pipeline(question=row["question"], context=row["context"])
    predictions.append(result["answer"])

df["predicted_answer"] = predictions
df
```

```python
def normalize_text(text):
    return str(text).strip().replace("ـ", "").replace("  ", " ")

df["answer_norm"] = df["answer"].apply(normalize_text)
df["predicted_norm"] = df["predicted_answer"].apply(normalize_text)

df["exact_match"] = (df["answer_norm"] == df["predicted_norm"]).astype(int)

accuracy = df["exact_match"].mean()
print("Exact Match Accuracy:", accuracy)
```
